```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

# LUAD LUSC Healthy

```python
df= pd.read_csv("../Data/luad_lusc_healthy_processed.csv")
df.index= df['Unnamed: 0']
df= df.iloc[:,2:]
df
```

```python
skf = StratifiedKFold(n_splits=5)
```

```python
df['label'].unique().tolist()
```

```python
target = df.loc[:,'label']
len(target)
```

```python
for train_index, test_index in skf.split(df, target):
    print(train_index)
```

```python
fold_no = 1
for train_index, test_index in skf.split(df, target):
    train = df.iloc[train_index,:]
    test = df.iloc[test_index,:]
    train_filename = 'train_split_' + str(fold_no) + '.csv'
    test_filename = 'test_split_' + str(fold_no) + '.csv'
    train.to_csv('../Data/' + train_filename, index=False)
    test.to_csv('../Data/' + test_filename, index=False)
    fold_no += 1
```
