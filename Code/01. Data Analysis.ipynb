# GDC Clinical Files

```python
luad_cli= pd.read_csv("../Data/gdc_clinical_luad.tsv", sep="\t")
luad_cli
```

```python
luad_cli.columns.tolist()
```

```python
luad_cli_s= luad_cli[['case_submitter_id', 'ethnicity', 'gender', 'race']]
luad_cli_s
```

```python
luad_ids= list(set(luad_cli_s['case_submitter_id']) & set(luaddf['Sample_2']))
luad_ids
```

```python
luad_cli_filtered= luad_cli_s[luad_cli_s['case_submitter_id'].isin(luad_ids)]
luad_cli_filtered
```

```python
luad_eam_ids= luad_cli_filtered[(luad_cli_filtered['ethnicity']=='not hispanic or latino') & (luad_cli_filtered['gender']=='male') & ((luad_cli_filtered['race']=='white'))]
pd.DataFrame(luad_eam_ids['case_submitter_id'].unique().tolist()).to_csv("luad_eam_ids.csv")
luad_eam_ids['case_submitter_id'].unique().shape
```

```python
luad_eaf_ids= luad_cli_filtered[(luad_cli_filtered['ethnicity']=='not hispanic or latino') & (luad_cli_filtered['gender']=='female') & ((luad_cli_filtered['race']=='white'))]
pd.DataFrame(luad_eaf_ids['case_submitter_id'].unique().tolist()).to_csv("luad_eaf_ids.csv")
luad_eaf_ids['case_submitter_id'].unique().shape
```

```python
luad_aam_ids= luad_cli_filtered[(luad_cli_filtered['ethnicity']=='not hispanic or latino') & (luad_cli_filtered['gender']=='male') & ((luad_cli_filtered['race']=='black or african american'))]
pd.DataFrame(luad_aam_ids['case_submitter_id'].unique().tolist()).to_csv("luad_aam_ids.csv")
luad_aam_ids['case_submitter_id'].unique().shape
```

```python
luad_aaf_ids= luad_cli_filtered[(luad_cli_filtered['ethnicity']=='not hispanic or latino') & (luad_cli_filtered['gender']=='female') & ((luad_cli_filtered['race']=='black or african american'))]
pd.DataFrame(luad_aaf_ids['case_submitter_id'].unique().tolist()).to_csv("luad_aaf_ids.csv")
luad_aaf_ids['case_submitter_id'].unique().shape
```

```python
combined_list_luad = [item for sublist in [luad_eam_ids['case_submitter_id'].unique().tolist(), luad_eaf_ids['case_submitter_id'].unique().tolist(), luad_aam_ids['case_submitter_id'].unique().tolist(), luad_aaf_ids['case_submitter_id'].unique().tolist()] for item in sublist]
len(combined_list_luad)
```

```python
lusc_cli= pd.read_csv("../Data/gdc_clinical_lusc.tsv", sep="\t")
lusc_cli_s= lusc_cli[['case_submitter_id', 'ethnicity', 'gender', 'race']]
lusc_ids= list(set(lusc_cli_s['case_submitter_id']) & set(luscdf['Sample_2']))
lusc_cli_filtered= lusc_cli_s[lusc_cli_s['case_submitter_id'].isin(lusc_ids)]
lusc_cli_filtered
```

```python
lusc_eam_ids= lusc_cli_filtered[(lusc_cli_filtered['ethnicity']=='not hispanic or latino') & (lusc_cli_filtered['gender']=='male') & ((lusc_cli_filtered['race']=='white'))]
pd.DataFrame(lusc_eam_ids['case_submitter_id'].unique().tolist()).to_csv("lusc_eam_ids.csv")
lusc_eam_ids['case_submitter_id'].unique().shape
```

```python
lusc_eaf_ids= lusc_cli_filtered[(lusc_cli_filtered['ethnicity']=='not hispanic or latino') & (lusc_cli_filtered['gender']=='female') & ((lusc_cli_filtered['race']=='white'))]
pd.DataFrame(lusc_eaf_ids['case_submitter_id'].unique().tolist()).to_csv("lusc_eaf_ids.csv")
lusc_eaf_ids['case_submitter_id'].unique().shape
```

```python
lusc_aam_ids= lusc_cli_filtered[(lusc_cli_filtered['ethnicity']=='not hispanic or latino') & (lusc_cli_filtered['gender']=='male') & ((lusc_cli_filtered['race']=='black or african american'))]
pd.DataFrame(lusc_aam_ids['case_submitter_id'].unique().tolist()).to_csv("lusc_aam_ids.csv")
lusc_aam_ids['case_submitter_id'].unique().shape
```

```python
lusc_aaf_ids= lusc_cli_filtered[(lusc_cli_filtered['ethnicity']=='not hispanic or latino') & (lusc_cli_filtered['gender']=='female') & ((lusc_cli_filtered['race']=='black or african american'))]
pd.DataFrame(lusc_aaf_ids['case_submitter_id'].unique().tolist()).to_csv("lusc_aaf_ids.csv")
lusc_aaf_ids['case_submitter_id'].unique().shape
```

```python
combined_list_lusc = [item for sublist in [lusc_eam_ids['case_submitter_id'].unique().tolist(), lusc_eaf_ids['case_submitter_id'].unique().tolist(), lusc_aam_ids['case_submitter_id'].unique().tolist(), lusc_aaf_ids['case_submitter_id'].unique().tolist()] for item in sublist]
len(combined_list_lusc)
```

```python
healthydf_filtered = healthydf[healthydf['Unnamed: 0'].str.contains('GTEX')]
list_healthy= healthydf_filtered['Unnamed: 0'].tolist()
len(list_healthy)
```

```python
# combining all sample IDs

combined_list = [item for sublist in [combined_list_luad, combined_list_lusc, list_healthy] for item in sublist]
len(combined_list)
```

```python
# keeping the first sample among the duplicates ids

df_uniqueSamples= df.drop_duplicates(subset='Sample', keep='first')
df_uniqueSamples
```

```python
# luad lusc samples

df_uniqueSamples_luad_lusc= df_uniqueSamples[~(df_uniqueSamples['label']=='healthy')]
df_uniqueSamples_luad_lusc
```

```python
# healthy samples only from GTEX

healthy_gtex= df_uniqueSamples[df_uniqueSamples['Unnamed: 0'].str.contains('GTEX')]
healthy_gtex
```

```python
# luad samples
luadSamples= df_uniqueSamples_luad_lusc[df_uniqueSamples_luad_lusc['Sample_2'].isin(combined_list_luad)]
luadSamples.shape
```

```python
# lusc samples
luscSamples= df_uniqueSamples_luad_lusc[df_uniqueSamples_luad_lusc['Sample_2'].isin(combined_list_lusc)]
luscSamples.shape
```

```python
healthy_gtex.shape
```

```python
combined_df_luad_lusc_healthy = pd.concat([luadSamples, luscSamples, healthy_gtex], ignore_index=True)

combined_df_luad_lusc_healthy
```

```python
combined_df_luad_lusc_healthy.to_csv("../Data/luad_lusc_healthy_processed.csv")
```
